```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]

x
y
```

```python
x.drop("model", axis=1, inplace=True)
x
```

```python
from sklearn.preprocessing import MinMaxScaler
# Min-Max scaling (Normalization) 
scaler_minmax = MinMaxScaler()  
scaler_data = scaler_minmax.fit_transform(x) 

# Print the scaled features using DataFrame 
x = pd.DataFrame(scaler_data,columns=x.columns )
x
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.6,random_state=21)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
train_score = regressor.score(x_train,y_train)
print(f'The Accuracy of training: {round((train_score*100),2)} %')

test_score = regressor.score(x_test,y_test)
print(f'The Accuracy of test: {round((test_score*100),2)} %')
```
